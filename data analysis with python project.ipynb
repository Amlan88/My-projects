#EDA#

```python
#plotting a bar chart 
```

```python
ax=sns.countplot(x= 'Gender', data=data)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#from the above bar graph it can be analysed that count of females are more than males
```

```python
gen_sales=data.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y= 'Amount' ,data= gen_sales )
```

```python
# from the above graph we can see that the females are purchasing more than men
```

```python
#Age#
```

```python
ax=sns.countplot(data=data,x='Age Group',hue= 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# from the above chart we can see that age group between 26-35 are most number of buyers and females are the most
```

```python
#Total number of orders from top 10 states
```

```python
sales_state = data.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(16,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Orders')
```

```python
## total amount/sales from top 10 states
```

```python
sales_state = data.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(16,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Amount')
```

```python
#from the above we can analyze that most of the orders&total sales/amount are from uttar pradesh,Maharashtra,Karnataka
```

```python
#Martial status# 
```

```python
ax= sns.countplot(data = data,x ='Marital_Status',hue='Gender')

sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#from the above graph we can analyze that most of the buyers are married(woman) and they have high purchasing power
```

```python
OCCUPATION
```

```python
sns.set(rc={'figure.figsize':(20,5)})
ax= sns.countplot(data = data,x = 'Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = data.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Occupation',y= 'Amount')
```

```python
from the above bar graph we can analyze that  most number of buyers are working in IT,healthcare,avaiation sector
```

```python
PRODUCT CATEGORY
```

```python
sns.set(rc={'figure.figsize':(27,5)})
ax= sns.countplot(data= data, x= 'Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = data.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(27,5)})
sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount')
```

```python
from the above graph we can analyze that most of the products are from food,clothing&electronics
```

```python
#Top 10 most sold products
```

```python
sales_state = data.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_ID',y= 'Orders')
```

```python
Conclusion:-
    from the above analysis we get that Married women age group 26-35 yrs from UP, Maharastra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from Food, Clothing and Electronics category
```
